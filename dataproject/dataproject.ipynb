```python
import pandas as pd
import matplotlib.pyplot as plt

# Opretter data
data = {
    'År': [2024, 2025, 2026, 2027, 2028, 2029],
    'Befolkning': [5_823_000, 5_875_000, 5_925_000, 5_978_000, 6_035_000, None]  # Den sidste værdi er tom og skal udfyldes
}

# Omdan datasættet til en Pandas dataframe
df = pd.DataFrame(data)

# Beregn befolkningstilvæksten baseret på en vækstrate på 1,59%
for i in range(1, len(df)):
    df.loc[i, 'Befolkning'] = df.loc[i-1, 'Befolkning'] * 1.0159

# Udskriv dataframe
print("Dataframe med befolkningstal:")
print(df)
print()

# Beregn statistikker
mean_population = df['Befolkning'].mean()
median_population = df['Befolkning'].median()
min_population = df['Befolkning'].min()
max_population = df['Befolkning'].max()

# Udskriv statistikker
print("Statistikker for befolkningstal:")
print(f"Gennemsnitlig befolkningstal: {mean_population:.2f}")
print(f"Median befolkningstal: {median_population:.2f}")
print(f"Minimum befolkningstal: {min_population:.2f}")
print(f"Maksimum befolkningstal: {max_population:.2f}")
print()

# Visualiser befolkningstilvæksten over tid
plt.figure(figsize=(10, 6))
plt.plot(df['År'], df['Befolkning'], marker='o', color='blue', linestyle='-')
plt.xlabel('År')
plt.ylabel('Befolkning')
plt.title('Forventet befolkningsvækst i Danmark')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Opretter data
data = {
    'År': range(2024, 2031),  # Årene fra 2024 til 2030
    'Befolkning': [None] * 7,  # Liste til at gemme forudsigede befolkningstal, starter som None
    'Levealder': [None] * 7  # Liste til at gemme forudsigede gennemsnitlige levealder, starter som None
}

# Omdan datasættet til en Pandas dataframe
df = pd.DataFrame(data)

# Antagninger
start_befolkning = 6024000  # Befolkningstallet i 2024
start_levealder = 80.5  # Gennemsnitlig levealder i 2024

# Definer en funktion for ikke-lineær vækst med tilfældige udsving
def ikke_lineær_vækst(år, startværdi, vækst_rate, stdev):
    return startværdi * np.exp(np.random.normal(vækst_rate, stdev) * (år - 2024))

# Forudsig befolkningstallet og den gennemsnitlige levealder for hvert år fra 2024 til 2030
for i in range(7):
    df.loc[i, 'Befolkning'] = ikke_lineær_vækst(2024 + i, start_befolkning, -0.05, 0.02)  # Tilfældige udsving med standardafvigelse på 0.02
    df.loc[i, 'Levealder'] = ikke_lineær_vækst(2024 + i, start_levealder, -0.02, 0.01)  # Tilfældige udsving med standardafvigelse på 0.01

# Lav en graf
plt.figure(figsize=(10, 6))
plt.plot(df['År'], df['Befolkning'], marker='o', color='blue', label='Befolkningstal')
plt.plot(df['År'], df['Levealder'], marker='o', color='red', label='Gennemsnitlig levealder')
plt.xlabel('År')
plt.ylabel('Værdi')
plt.title('Forudsagt befolkningstal og gennemsnitlig levealder fra 2024 til 2030 med tilfældige udsving')
plt.legend()
plt.grid(True)
plt.show()





```

# Merge data sets

```python
import pandas as pd
import matplotlib.pyplot as plt

# Danmark data
danmark_data = {
    'Alder': ['Under 20', '20-40', '40-60', '60-80', '80-100', '100-110'],
    'Danmark Population': [1287397, 1552284, 1522629, 1281138, 321332, 1210]
}

# Tyskland data
tyskland_data = {
    'Alder': ['Under 20', '20-40', '40-60', '60-80', '80-100', '100-110'],
    'Tyskland Population': [15859462.86, 20667917.03, 23029964.69, 18727663.59, 6073836.84, 843.58845]
}

# Omdan datasættet til en Pandas dataframe
df_danmark = pd.DataFrame(danmark_data)
df_tyskland = pd.DataFrame(tyskland_data)

# Beregn procentandelen af befolkningen i hver aldersgruppe for begge lande
df_danmark['Danmark Procent'] = (df_danmark['Danmark Population'] / df_danmark['Danmark Population'].sum()) * 100
df_tyskland['Tyskland Procent'] = (df_tyskland['Tyskland Population'] / df_tyskland['Tyskland Population'].sum()) * 100

# Lav en linjegraf for procentandelen af befolkningen i hver aldersgruppe for begge lande
plt.figure(figsize=(10, 6))

plt.plot(df_danmark['Alder'], df_danmark['Danmark Procent'], marker='o', label='Danmark', color='blue')
plt.plot(df_tyskland['Alder'], df_tyskland['Tyskland Procent'], marker='o', label='Tyskland', color='red')

plt.xlabel('Alder')
plt.ylabel('Procent af befolkningen')
plt.title('Sammenligning af aldersfordeling mellem Danmark og Tyskland')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotér x-aksens labels for bedre læsbarhed
plt.tight_layout()  # Justér layout for at undgå overlapning af labels
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Danmark data
danmark_data = {
    'Alder': ['Under 20', '20-40', '40-60', '60-80', '80-100', '100-110'],
    'Danmark Population': [1287397, 1552284, 1522629, 1281138, 321332, 1210]
}

# Tyskland data
tyskland_data = {
    'Alder': ['Under 20', '20-40', '40-60', '60-80', '80-100', '100-110'],
    'Tyskland Population': [15859462.86, 20667917.03, 23029964.69, 18727663.59, 6073836.84, 843.58845]
}

# Omdan datasættet til en Pandas dataframe
df_danmark = pd.DataFrame(danmark_data)
df_tyskland = pd.DataFrame(tyskland_data)

# Lav de demografiske pyramider
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Danmark
axs[0].barh(df_danmark['Alder'], df_danmark['Danmark Population'], color='blue')
axs[0].set_xlabel('Antal personer')
axs[0].set_ylabel('Alder')
axs[0].set_title('Demografisk pyramide - Danmark')

# Tyskland
axs[1].barh(df_tyskland['Alder'], df_tyskland['Tyskland Population'], color='red')
axs[1].set_xlabel('Antal personer')
axs[1].set_ylabel('Alder')
axs[1].set_title('Demografisk pyramide - Tyskland')

plt.tight_layout()
plt.show()
```
