```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def d(x) : return 1830 / (x+5) # Demand function
def s(x) : return 2 + 9 * x + x ** 2 # Supply function
```

```python
from scipy.optimize import fsolve
def f(xy) :
    x, y = xy
    z= np.array([y-d(x), y-s(x)])
    return z 
eq = fsolve(f, [0, 10])
x_e  = int(eq[0])
x_p  = int(eq[1])

print("x_e =", x_e) 
print("x_p =", x_p)
```

```python


x = np.linspace(0, 10, 10)
plt.plot(x, d(x), '-b', label= 'Demand function d(x)')
plt.plot(x, s(x), '-g', label= 'Supply function s(x)')

plt.axhline(y=x_p, color="Blue", linestyle="--")
plt.axvline(x=x_e, color="Blue", linestyle="--")

plt.plot(x_e, x_p, 'ro', label='Equilibrium point')

x = np.linspace(0, x_e, 1000)
plt.fill_between(x, d(x), x_p, color="Red", alpha=0.5)
plt.fill_between(x, x_p, s(x), color="Green", alpha=0.5)

plt.xlabel("Quantity")
plt.ylabel("Price")
plt.legend()

# Her angiver vi de ønskede tickværdier og etiketter på x og y-akserne
plt.xticks([0, x_e], ["0", "$x_e$"])
plt.yticks([0, x_p], ["0", "$x_p$"])


plt.show()
```

# Total Revenue
the Area of the red and green triangle is represented by the following:


# A = $x_e$ * $x_p$

```python
TR =  x_e * x_p 

print(TR)
```

# Consumer Surplus 
Consumer Surplus = $\int_{(0)}^{x_e} d(x) \, dx$ - TR

```python
x = sy.Symbol("x")
result1 = sy.integrate(d(x), (x, 0, x_e)) - TR

print("CS = ", "%.2f " %result1)
```

# Producer Surplus 

Producer Surplus =  TR - $\int_{(0)}^{x_e} s(x) \, dx$ 



```python
x = sy.Symbol("x")
result2 = TR - sy.integrate(s(x), (x, 0, x_e)) 

print("PS = ", "%.2f " %result2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Juster størrelsen på teksten i plottet
plt.rcParams["font.size"] = 8

# Funktioner for efterspørgsel og udbud
def d(x):
    return 1830 / (x + 5)  # Efterspørgselsfunktion

def s(x):
    return 2 + 9 * x + x ** 2  # Udbudsfunktion

# Definer x-værdier
x_values = np.linspace(0.1, 10, 1000)

# Plot efterspørgsels- og udbudsfunktioner
plt.plot(x_values, d(x_values), '-b', label='Demand functionn d(x)')
plt.plot(x_values, s(x_values), '-g', label='Supply function s(x)')

# Find ligevægtspunktet
x_e = None
for x in x_values:
    if abs(d(x) - s(x)) < 1:  # Ændret toleranceniveau til 1
        x_e = x
        break

# Plot ligevægtspunktet
if x_e is not None:
    plt.plot(x_e, d(x_e), 'ro', label='Equilibrium point')
else:
    print("NO Equilibrium found")

# Marker ligevægtspris og -mængde
if x_e is not None:
    plt.axhline(y=d(x_e), color="bLack", linestyle="--", label='Equilibrium price')
    plt.axvline(x=x_e, color="Grey", linestyle="--", label='Equilibrium quantity')

# Definer skat
tax = 20  # Lavere skat på efterspørgsel

# Ændre efterspørgselsfunktionen med skat
def d_tax(x):
    return d(x) - tax

# Plot efterspørgselsfunktionen med skat
plt.plot(x_values, d_tax(x_values), '-r', label='Demand function w tax d_tax(x)')

# Find nyt ligevægtspunkt med skat
x_e_tax = None
for x in x_values:
    if abs(d_tax(x) - s(x)) < 1:  # Ændret toleranceniveau til 1
        x_e_tax = x
        break

# Plot nyt ligevægtspunkt med skat
if x_e_tax is not None:
    plt.plot(x_e_tax, d_tax(x_e_tax), 'go', label='New Supply function w tax')

    # Marker nyt ligevægtspris og -mængde med skat
    plt.axhline(y=d_tax(x_e_tax), color="Yellow", linestyle="--", label='New Equilibrium price w tax')
    plt.axvline(x=x_e_tax, color="Orange", linestyle="--", label='New Equilibrium quantity w tax')

plt.xlabel("Quantity")
plt.ylabel("Price")
plt.legend()
plt.show()

# Udskriv ligevægtsmængde og ligevægtspris med og uden skat
if x_e is not None:
    print("Equilibrium quantity without tax =", round(x_e, 2))
    print("Equilibrium price without tax =", round(d(x_e), 2))
else:
    print("Ingen ligevægt fundet")

if x_e_tax is not None:
    print("Equilibrium quantity with tax =", round(x_e_tax, 2))
    print("Equilibrium price with tax =", round(d_tax(x_e_tax), 2))
else:
    print("NO Equilibrium found with tax")




```
